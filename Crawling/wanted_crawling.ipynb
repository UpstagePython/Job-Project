```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# ChromeDriverManager().install()
```

```python
browser = webdriver.Chrome()
```

```python
# 검색어 입력
search = input('타이틀을 입력: ')
url = f'https://www.wanted.co.kr/search?query={search}&tab=position'
browser.get(url)
```

```python
from selenium.webdriver.common.by import By
import time

# 반복을 통해 각 요소 클릭 및 크롤링
job_data = []

cnt = int(browser.find_element(By.XPATH, '//*[@id="search_tabpanel_position"]/div/div[1]/h2/span').text)
print(f'총 {cnt}개의 데이터를 수집합니다.')

for i in range(1, cnt+1):
    # 요소 클릭
    browser.find_element(By.XPATH, f'//*[@id="search_tabpanel_position"]/div/div[4]/div[{i}]/a/div[1]').click()
    # 시간 지연
    time.sleep(5)
    
    # 요소 가져오기
    job_header = browser.find_element(By.CLASS_NAME, 'JobHeader_JobHeader__DBrFJ')
    job = job_header.text

    # \n을 기준으로 분리
    job_info = job.split('\n')

    # 직무명
    title = job_info[1]

    # ∙을 기준으로 분리
    title_info = job_info[0].split('∙')

    # 회사명, 회사 위치, 경력 추출
    company = title_info[0].strip()
    region = title_info[1].strip()
    career = title_info[2].strip()

    # 마감 기한 정보 가져오기
    deadline_element = browser.find_element(By.CLASS_NAME, 'JobDueTime_JobDueTime__iKbEO')
    deadline = deadline_element.text.split('\n')[1].strip()

    # 현재 링크 가져오기
    link = browser.current_url

    # 데이터 저장
    job_data.append({
        'company': company,
        'region': region,
        'career': career,
        'title': title,
        'deadline': deadline,
        'link': link
    })
    
    print(f'{company}의 {title} 데이터를 수집했습니다. ({i}/{cnt})')

    # 뒤로 가기
    browser.back()
    
    # i가 4의 배수일 때 스크롤을 50 내림
    if i % 4 == 0:
        # 현재 스크롤 위치 및 페이지 높이 확인
        scroll_position = browser.execute_script("return window.pageYOffset;")
        page_height = browser.execute_script("return document.body.scrollHeight;")
        
        # 스크롤을 더 할 수 있는지 확인
        if scroll_position + browser.execute_script("return window.innerHeight;") < page_height:
            browser.execute_script("window.scrollBy(0, 50);")
            time.sleep(1)

job_data
```
